### Diccionarios por comprensi贸n

De forma an谩loga a como se escriben las listas por comprensi贸n, podemos aplicar este m茅todo a los *diccionarios por comprensi贸n*:

```python
for letter, count in letter_counts.items():
    print(letter, count)
```

> Una mejora de esta comprensi贸n ser铆a usar un conjunto para evitar contar letras repetidas:  
`{letter: word.count(letter) for letter in set(word)}`

Tambi茅n podemos incorporar *condiciones* en un diccionario por comprensi贸n:

```python
vowels = 'aeiou'
word = 'onomatopoeia'
```

```python
vowel_counts = {letter: word.count(letter)
                for letter in set(word)
                if letter in vowels}
```

```python
for vowel, count in vowel_counts.items():
    print(vowel, count)
```
