### Reunir/Desplegar argumentos posicionales

Python ofrece la posibilidad de utilizar un asterisco `*` en los par谩metros de las funciones. Sirve para **reunir** m煤ltiples argumentos posicionales en una 煤nica tupla como valor del par谩metro.

```python
def print_args(*args):
    print('Positional tuple:', args)
```

Tambi茅n podemos utilizar esta estrategia para establecer en una funci贸n una serie de par谩metros como *requeridos* y recibir el *resto de argumentos* como opcionales y empaquetados:

##  Generadores

Un **generador** es un objeto que premite crear secuencias. La gran ventaja de usar generadores es que podemos iterar sobre enormes secuencias sin necesidad de crearlas ni de almacenarlas completamente en memoria de una sola vez.

Los generadores suelen ser la fuente de datos de los *iteradores*. De hecho ya hemos usado una de ellos, `range()`, para generar secuencias de valores enteros:

```python
sum(range(1, 101))
```

> Cada vez que iteramos a trav茅s de un generador se lleva un seguimiento del 煤ltimo valor generado para poder generar el siguiente (si procede). Esto es diferente de una funci贸n ordinaria, que no tiene "memoria" de sus llamadas anteriores y siempre empieza desde la primera l铆nea con el mismo estado.

### Funciones generadoras

Si necesitamos crear una secuencia *potencialmente larga* podemos escribir una **funci贸n generadora**. Se trata de una funci贸n ordinaria pero que retorna su valor con `yield` en vez de con `return`.

Veamos un ejemplo en el que escribimos nuestra propia versi贸n de `range()`:

```python
def my_range(first=0, last=10, step=1):
    number = first
    while number < last:
        yield number
        number += step
```

```python
my_range  # es una funci贸n ordinaria
```

```python
ranger = my_range(1, 5)  # devuelve un generador
ranger
```

Podemos iterar sobre este objeto generador:

```python
for i in ranger:
    print(i)
```

Un detalle muy importante sobre los generadores es que "*se agotan*". Es decir, una vez que ya hemos consumido todos sus elementos ya no obtendremos nuevos valores:

```python
for i in ranger:
    print(i)
```

### Expresiones generadoras

Una **expresi贸n generadora** es sint谩cticamente muy similar a una lista por comprensi贸n, pero utilizamos par茅ntesis en vez de corchetes. Se podr铆a ver como *una versi贸n acortada de una funci贸n generadora*.

```python
genobj
```

```python
for even_number in genobj:
    print(even_number)
```

##  Decoradores

Hay veces que necesitamos modificar una funci贸n existente sin cambiar su c贸digo fuente. Un ejemplo muy com煤n es a帽adir algunas sentencias de depuraci贸n para ver qu茅 argumentos estamos pasando.

Un **decorador** es una *funci贸n que toma como entrada una funci贸n y devuelve otra funci贸n*.

Veamos un ejemplo en el que documentamos la ejecuci贸n de una funci贸n:

```python
def document_it(func):
    def new_function(*args, **kwargs):
        print('Running function:', func.__name__)
        print('Positional arguments:', args)
        print('Keyword arguments:', kwargs)
        result = func(*args, **kwargs)
        print('Result', result)
        return result
    return new_function
```

Ahora definimos una funci贸n ordinaria que suma dos valores por par谩metros:

```python
def add_ints(a, b):
    return a + b
```

```python
add_ints(3, 5)
```

Hasta aqu铆 todo normal. Pero si aplicamos nuestro *decorador* a la funci贸n `add_ints` obtendremos otra funci贸n "*decorada*" que nos muestra m谩s informaci贸n por pantalla al ejecutarla:

```python
documented_add_ints = document_it(add_ints)
documented_add_ints(3, 5)
```

### Usando `@` para decorar

Como una alternativa a la aplicaci贸n manual de un decorador podemos usar `@` (seguido del nombre del decorador) antes de la definici贸n de la funci贸n que queremos decorar:

```python
@document_it
def add_ints(a, b):
    return a + b
```

```python
add_ints(3, 5)
```

Podemos aplicar *m谩s de un decorador* a cada funci贸n. Vamos a crear otro decorador que eleva al cuadrado el resultado:

```python
def square_it(func):
    def new_function(*args, **kwargs):
        result = func(*args, **kwargs)
        return result * result
    return new_function
```

El decorador que est谩 m谩s cerca de la funci贸n se ejecuta primero...

```python
@document_it
@square_it
def add_ints(a, b):
    return a + b
```

```python
add_ints(3, 5)
```

## 伙 Recursividad

La **recursividad** es el mecanismo por el cual una funci贸n se llama a s铆 misma.

```python
def dive():
    return dive()
```

```python
dive()
```

> Python muestra un error si no controlamos las llamadas recursividad.

Un ejemplo claro de funci贸n recursiva es el *c谩lculo del en茅simo t茅rmino* de la [sucesi贸n de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci).

Veamos una posible implementaci贸n en Python:

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
fibonacci(10)
```

```python
fibonacci(20)
```

Ya que hemos visto las *funciones generadoras*, parece claro que podr铆amos adaptar el ejemplo anterior de *Fibonacci* y aplicarlas aqu铆 (junto con una funci贸n interior):

```python
def fibonacci():
    def fibonacci_helper(n):
        if n == 0:
            return 0
        if n == 1:
            return 1
        return fibonacci_helper(n - 1) + fibonacci_helper(n - 2)

    n = 0
    while True:
        yield fibonacci_helper(n)
        n += 1
```

```python
fib = fibonacci()       # creaci贸n del generador
for _ in range(10):  # diez primeros t茅rminos
    print(next(fib))
```

##  Excepciones

Una **excepci贸n** es el c贸digo que se ejecuta cuando se produce un error en nuestro programa Python durante la fase de ejecuci贸n.

De hecho ya hemos visto algunas de estas excepciones: accesos fuera de rango a listas o tuplas, accesos a claves inexistentes en diccionarios, etc. Cuando ejecutamos c贸digo que podr铆a fallar bajo ciertas circunstancias necesitamos tambi茅n *manejar las excepciones* de manera adecuada.

Si una excepci贸n ocurre en una funci贸n y no es capturada en ese punto, va subiendo (*burbujeando*) hasta que es capturada en alguna funci贸n que ha hecho la llamada. Si en toda la "pila" de llamadas no existe un control de la excepci贸n, Python muestra un mensaje de error con informaci贸n adicional.

```python
short_list = ['a', 'b', 'c']
short_list[5]
```

### Manejando errores

Para manejar (*capturar*) las excepciones podemos usar un bloque de c贸digo con las palabras reservadas `try` and `except`:

```python
short_list = ['a', 'b', 'c']
position = 5
try:
    short_list[position]
except:
    print(f'Need a position between 0 and {len(short_list) - 1} but got {position}')
```

El c贸digo que est谩 dentro del bloque `try` se ejecuta, si hay un error se "*levanta*" una excepci贸n y se ejecuta el c贸digo que est谩 dentro del bloque `except`. En caso de que no hayan errores no se ejecuta el bloque `except`.

> No es una buena pr谩ctica usar un bloque `except` sin indicar el tipo de excepci贸n que estamos gestionando, no s贸lo porque puedan existir varias excepciones sino porque "expl铆cito" es mejor que "impl铆cito".

En el siguiente ejemplo mejoraremos el c贸digo anterior capturando en primer lugar `IndexError` y luego capturando cualquier otra excepci贸n que se pueda producir:

```python
short_list = ['a', 'b', 'c']
while True:
    value = input('Position [q to quit]? ')
    if value == 'q':
        break
    try:
        position = int(value)
        print(short_list[position])
    except IndexError as err:
        print('Bad index:', position)
    except Exception as other:
        print('Something else broke:', other)
```

### Crear excepciones propias

Todas las excepciones que hemos visto hasta ahora estaban ya predefinidas en el propio lenguaje o en la librer铆a est谩ndar. Pero es posible crear *excepciones propias* para manejar situaciones especiales que podr铆an producirse en nuestro c贸digo:

```python
class UppercaseException(Exception):
    pass
```

```python
words = ['chocolate', 'milk', 'TEA', 'water']
```

```python
for word in words:
    if word.isupper():
        raise UppercaseException(word)
```

> Esta excepci贸n propia tambi茅n se puede gestionar con un `try...except`.
